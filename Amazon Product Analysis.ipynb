# 4. Data Cleaning and Transformation

### 4.1 Examine for any null values in the dataset

```python
# Lets start by finding the null values in the columns 
df.isnull().sum()
```

### 4.4 Fixing Unreadable values 

```python
#Identify and Decode the Column
print(df['actual_price'].unique())
```

```python
df['actual_price'] = df['actual_price'].str.encode('latin1').str.decode('utf-8')
```

```python
df['actual_price'] = df['actual_price'].str.replace('₹', '', regex=False).str.strip()
```

```python
df['actual_price'] = pd.to_numeric(df['actual_price'], errors='coerce')
```

```python
df['actual_price'].fillna(0, inplace=True)
```

```python
df.dropna(subset=['actual_price'], inplace=True)
```

```python
print(df['actual_price'].head())
print(df['actual_price'].dtype)  # Should now be 'float64' or 'int64'
```

```python
df.head()
```

```python
# Decode incorrectly encoded values
df['discounted_price'] = df['discounted_price'].str.encode('latin1').str.decode('utf-8')

# Remove the currency symbol and clean whitespace
df['discounted_price'] = df['discounted_price'].str.replace('₹', '', regex=False).str.strip()

# Convert to numeric values
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')

# Handle missing or invalid values
df['discounted_price'].fillna(0, inplace=True)
```

```python
df.dropna(subset=['discounted_price'], inplace=True)
```

```python
print(df['discounted_price'].head())
print(df['discounted_price'].dtype)
```

```python
df.head()
```

# 5. Data Visualization

### 5.1 Line Plot

Line plots are a powerful tool for visualizing the flow of data, especially when tracking continuous changes, comparing multiple datasets, or analyzing trends over time. In this context, a line plot can effectively illustrate the gap between actual and discounted prices, allowing us to identify products with significant discounts or price fluctuations. This insight can help in identifying pricing strategies or opportunities for further analysis and decision-making.

```python
sns.lineplot(data=df[['actual_price', 'discounted_price']])
plt.title('Line Plot for Actual Price vs Discounted Price')
plt.xlabel('Index')
plt.ylabel('Price')
plt.show()
```

The graph is hard to interpret due to the high number of products. To improve readability and gain clearer insights, we will focus on analyzing the top 15 highest-rated products

```python
# Sort the products based on rating in descending order to get the top 15 rated products
top_15_rated_products = df.sort_values(by='rating', ascending=False).head(15)

# Plotting the line graph
plt.figure(figsize=(12, 8))

# Plot Actual Prices
sns.lineplot(x='product_name', y='actual_price', data=top_15_rated_products, label='Actual Price', marker='o')

# Plot Discounted Prices
sns.lineplot(x='product_name', y='discounted_price', data=top_15_rated_products, label='Discounted Price', marker='o')

# Adding titles and labels
plt.title('Actual vs Discounted Prices of Top 15 Rated Products')
plt.xlabel('Product Name')
plt.ylabel('Price')
plt.xticks(rotation=45, ha='right')  # Rotate product names for better readability
plt.legend(title='Price Type')

# Display the plot
plt.tight_layout()
plt.show()
```

### Analysis:

- Most products in the chart are clustered around the lower price range (below 20,000), suggesting affordability as a key factor for these items.
- Products with steep discounts on higher actual prices (like the TV or the Xiaomi Pad) may act as flagship offers to draw consumer attention.
- Lower discounts for affordable items could be positioned as "everyday low prices".
- The overall trend line shows more substantial price reductions for premium products compared to mid-range or low-end items.

### 5.2 Bar plot
Bar plots are ideal for comparing quantities across different categories and highlighting trends in data that are not sequential or time-based. They are particularly effective for displaying aggregated data, such as totals or averages, making them a versatile tool for visualizing categorical information. In this scenario, a bar plot is well-suited for showcasing products based on their ratings while also illustrating how discounts influence the sale of these products, providing insights into customer preferences and the effectiveness of pricing strategies.

```python
# Sort the products based on rating in descending order to get the top 15 rated products
top_15_rated_products = df.sort_values(by='rating', ascending=False).head(15)

# Set the plot size
plt.figure(figsize=(12, 8))

# Create a bar plot with actual and discounted prices for each product
sns.barplot(x='actual_price', y='product_name', data=top_15_rated_products, label='Actual Price', color='blue', alpha=0.6)
sns.barplot(x='discounted_price', y='product_name', data=top_15_rated_products, label='Discounted Price', color='orange', alpha=0.6)

# Adding titles and labels
plt.title('Actual vs Discounted Prices of Top 15 Rated Products')
plt.xlabel('Price')
plt.ylabel('Product Name')
plt.legend(title='Price Type')

# Display the plot
plt.tight_layout()
plt.show()
```

### Analysis:

- The Sony Bravia 164 cm 4K Ultra HD Smart LED TV (KD-65X74K) has the highest price among the listed products with a substantial difference between the actual and discounted price.
-  Items like the Amazon Basics Wireless Mouse and Syncwire LTG USB Cable have relatively small price differences, indicating lower discounts.
-  Products with significant discounts, especially high-value items like the Sony Bravia TV, may attract more customers due to the perceived value for money.
-  Marketers could focus promotions on items with the most significant discounts to drive sales while emphasizing affordability on lower-priced items.

### Recommendations:
- Highlight products with the highest ratings (e.g., Amazon Basics Wireless Mouse) in marketing campaigns to leverage customer trust.
- Address variability in ratings for products with larger error bars by analyzing customer feedback to identify areas for improvement.

#### Moving forward, we delve into analyzing the top 10 highest-rated products to gain insights into their performance and customer feedback

```python
# Sort the data by rating in descending order and select the top 10
top_10_products = df.sort_values(by='rating', ascending=False).head(10)

# Plotting the graph
plt.figure(figsize=(10, 6))
sns.barplot(x='rating', y='product_name', data=top_10_products, hue='rating', legend=False)

# Adding titles and labels
plt.title('Top 10 Highest Rated Products')
plt.xlabel('Rating')
plt.ylabel('Product Name')

# Display the plot
plt.show()
```

### Analysis:

- All products in this visualization have ratings close to the top of the scale (4.7–5.0), indicating strong customer satisfaction across the board.
- While most products cluster around high ratings, the error bars indicate variability in customer reviews. Some products, despite high ratings, may have mixed feedback, which could be a result of individual customer experiences

### Recommendations:

- Highlight products with the highest ratings (e.g., Amazon Basics Wireless Mouse) in marketing campaigns to leverage customer trust.
- Address variability in ratings for products with larger error bars by analyzing customer feedback to identify areas for improvement.
